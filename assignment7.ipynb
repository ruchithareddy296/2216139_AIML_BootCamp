```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
dataset="/content/drive/MyDrive/SUV_Purchase.csv"
df=pd.read_csv(dataset)
df.head()
```

```python
#need to change the gender into 0,1
#data processing 
from sklearn import preprocessing 
#labelencodingg
LE=preprocessing.LabelEncoder() 
#fitting it to our dataset 
df.Gender=LE.fit_transform(df.Gender)
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df.Age,df.EstimatedSalary)
plt.show()
```

```python
import seaborn as sns
def get_heatmap(df):
  plt.figure(figsize=(6,4))
  sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
  plt.tight_layout()
  return plt.show()
get_heatmap(df)
```

```python
import seaborn as sns
with sns.plotting_context("notebook",font_scale=2):
  g=sns.pairplot(df[['User ID','Gender','Age','EstimatedSalary']],palette='tab20',size=4)
  g.set(xticklabels=[])
```

```python
from matplotlib.axis import YTick
#method 2 to load the data in the form of arrays -by library numpy
import numpy as np
X=np.array(df[['Gender','EstimatedSalary','Age']])
Y=np.array(df[['Purchased']])
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_train
```

```python
y_train
```

```python
y_pred=model.predict(x_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
print("Predicted value for training data:",model.score(x_train,y_train)*100)
```

```python
print("Predicted value for testing data:",model.score(x_test,y_test)*100)
```

```python
accuracy=model.score(X,Y)*100
print(accuracy)
```
